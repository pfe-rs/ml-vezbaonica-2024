## Uvod u pandas

##### Učitavanje potrebnih biblioteka

```python
import pandas as pd
import numpy as np
```

##### Učitavanje CarPrice_Assignment.csv u dataframe
Baza koja će se analizirati je - Analiza promene cene automobila (https://www.kaggle.com/datasets/hellbuoy/car-price-prediction)    


```python
car_price_df = pd.read_csv('CarPrice_Assignment.csv')

car_price_df
```

##### Izbacivanje određene kolone

```python
car_price_df.drop(['symboling', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation','wheelbase',
 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio',
 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg'], axis=1)
```

##### Sabiranje svih vrednosti u određenoj koloni

```python
car_price_df['price'].sum()
```

##### Maksimalna vrednost cene automobila

```python
car_price_df['price'].max()
```

##### Odrediti srednju cenu automobila i standardno odstupanje 

```python
car_price_df['price'].mean()
```

```python
car_price_df['price'].std()
```

##### Izdvajanje podataka o automobilima kojima je cena manja od 5500

```python
car_price_df.loc[car_price_df['price'] < 5500.0]
```

```python
car_price_df[car_price_df['price'] < 5500.0]
```

##### Primena linearne transformacije na celu kolonu 

```python
car_price_df['price'] = (car_price_df['price'] + 2)/2
car_price_df
```

##### Naći informacije o svim kolima modela volvo 246

```python
car_price_df[car_price_df['CarName'] == "volvo 144ea"]
```

##### Primeniti matematičke operacije na celu kolonu

```python
car_price_df['price'] = car_price_df['price'].apply(np.sqrt) 
car_price_df
```

```python
car_price_df['price'] = car_price_df['price'].apply(lambda x : x*x + 12)
car_price_df
```

##### Prikaz cene automobila na histogramu

```python
car_price_df['price'].hist()
```

##### Broj instanci svake marke automobila

```python
car_price_df['CarName'].value_counts()
```

##### Promena vrednosti na određenim pozicijama u dataFrame-u

```python
car_price_df.loc[0:3, ['wheelbase', 'price']] = 50
car_price_df
```

##### Dodavanje novog reda u dataFrame

```python
car_price_df.loc['novi_red'] = 0
car_price_df
```

##### Resetovanje svih cena automobila na nulu

```python
car_price_df.loc[:, 'price'] = 0
car_price_df
```

##### Sortiranje dataFramea po ceni automobila

```python
#because we've set the price to 0
car_price_df = car_price_df.sort_values(by = 'highwaympg', ascending = False)
car_price_df 
```

##### Opis statistike dataFrame-a

```python
df_new = car_price_df.describe()
df_new
```

##### Nadovezivanje dva dataFrame-a

```python
df = pd.concat([df_new, car_price_df])
df
```

##### Čuvanje novog dataFramea 

```python
df.to_csv("novi.csv", index = False)
```
